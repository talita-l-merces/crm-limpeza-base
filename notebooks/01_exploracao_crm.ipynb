```python
import re #Biblioteca padrão de expressões regulares (regex)
import pandas as pd
import unicodedata #Biblioteca padrão do Python para lidar com unicode(acentos,caracteres especiais) 
 
```

```python
#Função para limpar nome de coluna
# re.sub (padrão, substituto, texto) "Procure um padrão dentro do texto e substitua por outra coisa"
def limpar_nome_coluna(col:str) -> str:  #Diz que a coluna é texto e que a função devolve texto
    col=str(col).strip() #Reforço que a coluna é string e uso .strip para remover espaços
    col=re.sub(r"(?<=[a-z0-9])(?=[A-Z])"," ", col) #Este é um padrão mais avançado, ele insere um espaço entre letra minúscula e número e a outra entre letra maiúscula e número. Sintaxe, r significa: raw string
    # diz, náo interprete \ como espaço " ". Lookbehind (?<=...) "verifique o que vem antes mas não consuma, A a Z e 0 a 9. Lookhead "verifique o que vem depois mas nõs consuma"[A-Z] qualquer letra maiúscula 
    # existe qualquer letra maiúscula depois?. Até esse ponto somente identificação por fim após o parenteses o re.sub fala para substituir a vírgula por espaço
    col=re.sub(r"[_\-\/]+", " ", col)  #Pega _ ou - ou/ e troca por espaço, é usada a barra invertida para remover a ambiguidade e demonstrar que aquele é só o sómbolo mesmo.
    col=re.sub(r"\s+", " ", col). strip() # \s+ significa "um ou mais espaços ou tabs" troca tudo por um espaço só 
```

```python
## 3. Padronização de texto interno (strings)
#Objetivo: Remover espaços invisíveis, múltiplos espaços e ruídos de formação sem alterar os significado dos dados
def limpar_texto(x): #X é um placeholder, indica um valor qualquer vindo de uma célula do dataframe, ele precisa existir para ser o agente da iteração, a limpeza de texto passará por todos os valores do dataframe
    if pd.isna(x): # isna mean ïs not available" se o valor não estiver disponível ele retorna Nan ou None e não limpa pois não tem valor ali 
        return x
    s= str(x) # Indicamos que a partir de agora vamos trabalhar com texto
    s=s.replace("\u00a0", " ") # Este é um caractere unicode chamado de Non-breaking space, seria trocar o espaço da máquina por um espaço normal
    s=re.sub(r"\s+"," ",s) # Isso significa qualquer sequencia de espaços várias vezes
    return s.strip() # Retorna o s limpo e o último strip garante a limpeza das bordas
```

```python
import pandas as pd
import numpy as np

tel=(
    df_rd[telefone_col]
    .astype("string")
    .str.strip()
    .str.replace(r"\D+","", regex=True)
    .replace("", pd.NA)
)


```

```python
def padroniza_tel_com_ddi(x):
    if pd.isna(x):
        return pd.NA
    
    if len(x) in (10,11):
        return "55" + x

    if x.startswith("55") and len(x) in (12,13):
        return x

    if 8<= len(x) <=15:
        return x

    return pd.NA

df_rd["telefone_limpo"]= tel.apply(padroniza_tel_com_ddi)

df_rd["telefone_limpo"].str[:2].value_counts(dropna=False).head(10)
```

```python
df_rd["chave_contato"]=df_rd["email_limpo"]
df_rd.loc[df_rd["chave_contato"].isna(), "chave_contato"] = df_rd["telefone_limpo"]

df_rd["score_completude"]= df_rd.notna().sum(axis=1)
df_rd=df_rd.sort_values(by=["chave_contato", "score_completude"], ascending=[True,False])

df_dedupe = (
    df_rd
    .dropna(subset=["chave_contato"])
    .drop_duplicates(subset=["chave_contato"], keep="first")
)

print("Antes:", len(df_rd))
print("Depois(dedupe):", len(df_dedupe))
```
